```python
import requests
import pandas as pd

r = requests.get("http://0.0.0.0:5000/log").json()
logs = pd.DataFrame(r)

logs_sum = logs.groupby('acc_id').amount.sum()
logs
```

```python
r = requests.get("http://0.0.0.0:5000/account").json()
accs = pd.DataFrame(r)

accs.start_date = pd.to_datetime(accs.start_date)
accs.end_date = pd.to_datetime(accs.end_date)
accs["course_duration"] = accs.end_date - accs.start_date

accs.value = accs.id.apply(lambda x: logs_sum[x])
accs
```
